```python
import glob
import pandas as pd
import numpy as np
```

```python
file_names = glob.glob('*.csv')
print(file_names)
```

```python
file_names[0]
```

```python
data = pd.read_csv(file_names[0])
```

```python
data.head()
```

```python
data['metric'] = data['numerator'] / data['denominator']
```

```python
def pop_std(x):
    return x.std(ddof=0)
data.groupby(['metric_id', 'groupid']).agg({'metric': ['mean', 'std', pop_std, 'min', 'median', 'max'], 'uin': ['count', 'nunique']})
```
