```python
import re
import os
import numpy as np
import pandas as pd

fs_all = []
result_list = []
Result = []
algo = ['AB', 'EG', 'Elimination', 'SF', 'esSR', 'PSI', 'TS', 'UCB', 'US']
# Get the file list
file_path = './Exp_multiArm'
fs = os.listdir(file_path)
fs.sort()
algo_name = fs.copy()
for i in range(len(fs)):
    fs[i] = file_path + '/' + fs[i]
    fs_fs = os.listdir(fs[i])
    fs_fs.sort()
    for j in range(len(fs_fs)):
        fs_fs[j] = fs[i] + '/' + fs_fs[j]
    fs_all.append(fs_fs)

variance_index = [4, 5, 6, 7, 10, 11, 12]
dimension_index = [3, 8, 0, 1, 2]
arm_index = [15, 9, 13, 14]
index_all = [variance_index, dimension_index, arm_index]
index_name = ['variance', 'dimension', 'arm']
power_pattern = re.compile(r'Power=([\d\.]+)')
obs_pattern = re.compile(r'arm = ([\d\.]+)')
regret_pattern = re.compile(r'Regret = ([\d\.]+)')
pattern_all = [power_pattern, obs_pattern, regret_pattern]
head = ['AB_Test',
 'EG',
 'Elimination',
 'Elimination_SF',
 'PSI',
 'TS',
 'UCB',
 'US',
 'esSR']

with open('score.csv','w',encoding='utf8',newline='') as f_csv :
    for control_index in range(len(index_all)):
        f_w = open('Result_' + index_name[control_index] + '.txt', 'a')
        result_all_Power = []
        result_all_Obs = []
        result_all_Regret = []
        for algo_fs in range(len(fs_all)):
            #f_w.write(algo_name[algo_fs] + '\n')
            result_list_power = []
            result_list_obs = []
            result_list_regret = []
            for index in index_all[control_index]:
                with open(fs_all[algo_fs][index], 'r') as f:
                    content = f.read()
                    result_list_power.append(round(float((pattern_all[0].findall(content))[-1]), 3))
                    result_list_obs.append(round(float((pattern_all[1].findall(content))[-1]), 3))
                    result_list_regret.append(round(float((pattern_all[2].findall(content))[-1]), 3))
            result_all_Power.append(result_list_power)
            result_all_Obs.append(result_list_obs)
            result_all_Regret.append(result_list_regret)
        result_all_Power = np.array(result_all_Power)
        result_all_Obs = np.array(result_all_Obs)
        result_all_Regret = np.array(result_all_Regret)
        result_all_Power = result_all_Power.T
        result_all_Obs = result_all_Obs.T
        result_all_Regret = result_all_Regret.T
        f_w.write('Power: ' + '\n' + str(result_all_Power) + '\n')
        f_w.write('Obs: ' + '\n' + str(result_all_Obs) + '\n')
        f_w.write('Regret: ' + '\n' + str(result_all_Regret) + '\n')


        #f_w.write('Power: ' + str(result_list_power) + '\n')
        #f_w.write('Obs: ' + str(result_list_obs) + '\n')
        #f_w.write('Regret: ' + str(result_list_regret) + '\n')
```
